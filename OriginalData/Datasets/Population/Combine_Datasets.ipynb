```python
import pandas as pd
import os
link ='C:\\Users\\Maria\\Documents\\AESOPS\\Datasets\\Population\\'
os.chdir(link)
data = [i for i in os.listdir('ppn')]
```

```python
dt = pd.DataFrame()
data = [link+'ppn\\'+i for i in data]
y =0
```

```python
for i in data:
    x = i.split('\\')[-1].split('_')[0]
    df = pd.read_csv(i)
    df['County'] = x
    df['shape'] = df.shape[0]
    #print(dt.shape)
    dt = pd.concat((df,dt))
    #print(dt.shape)
    #print(df.head())
    y+=1
dt.to_csv('data.csv')
```
