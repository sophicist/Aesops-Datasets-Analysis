```python
import numpy as np
import pandas as pd
from ipywidgets import  *
import matplotlib.pyplot as plt
```

```python

df = pd.DataFrame(np.random.randint(0,100,(100,1)),columns =["stud"])
df["Height"] = np.random.randint(300,1000,(100,1))
df.head()
```

```python
def plotta(data ,color,title,x_var,y_var):
    plt.figure()
    plt.plot(data,color =color)
    plt.xlabel(x_var)
    plt.ylabel(y_var)
    plt.title(title)
    plt.show()
plotta(df.stud,'red',"thegere","khahahah","muhaha")
```

```python
interactive(plotta,data = df.stud,
         color = ["yellow","green","blue","turquoise","teal","cyan"],
        title = "faga",
        x_var = "fagu",
        y_var = "silly")
```

```python
def simple(x,color):
    plt.plot(x,color =color)
    plt.show()
simple(df.stud,"cyan")
```

```python
interact(simple,x = df.stud,color = ["green","blue"])
```
