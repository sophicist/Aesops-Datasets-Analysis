```python
#!pip install gapminder
from gapminder import gapminder as dx
import numpy as np
import os
import pandas as pd
df = pd.read_csv("gapminder_tidy.csv")
df.head()
```

```python

df["life"] = [np.log(i) for i in df["life"]]
df["gdp"] =np.log(df.gdp)
dx.head()
```

```python
subset = df[[True if i in [1970,2007] else False for i in df.Year ]]
subset.head()
```

# Matplotlib

```python
import matplotlib.pyplot as plt
sub  = subset[subset.Year == 1970]
plt.style.use("ggplot")
plt.figure(figsize =(10,7))
plt.scatter(sub["life"],sub["gdp"])
plt.title("Life expectancy vs Gdp in Gapminder 1970")
plt.ylabel("GDP in log(million $)")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize =(10,7))
sns.scatterplot(data =sub,x= "life",
                y = "gdp",size = "population",hue ="region")
plt.title("Life expectancy vs Gdp in Gapminder 1970")
plt.show()
```
